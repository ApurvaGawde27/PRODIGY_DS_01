```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("shopping-data.csv")
df.head(2)
```

```python
fig, ax = plt.subplots(figsize = (6,4))
df1=df['Annual Income (k$)']
df1.plot(kind='hist',bins=20)
ax.set_xlabel("Annual income")
ax.set_title("Annual income of customers")
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (6,4))
plt.bar(df['Annual Income (k$)'], df['Spending Score (1-100)'])
ax.set_xlabel("Annual income")
ax.set_ylabel("Spending Score")
plt.show()
```
